```python
!pip install opencv-python umap-learn hdbscan

import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, utils
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
import umap.umap_ as umap
import hdbscan
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
from hdbscan import approximate_predict

# ----------------------------------------------------
# 1. Load & Filter CIFAR-10 (Classes 0,1,8)
# ----------------------------------------------------
(x_train_full, y_train_full), (x_test_full, y_test_full) = tf.keras.datasets.cifar10.load_data()
selected_classes = [0, 1, 8]  # airplane=0, automobile=1, ship=8

train_mask = np.isin(y_train_full.squeeze(), selected_classes)

x_train = x_train_full[train_mask].astype('float32') / 255.0
y_train = y_train_full[train_mask].squeeze()
x_test = x_test_full.astype('float32') / 255.0

# Remap labels => {0 -> 0, 1 -> 1, 8 -> 2}
label_map = {0:0, 1:1, 8:2}
y_train_mapped = np.array([label_map[lab] for lab in y_train])
y_train_oh = utils.to_categorical(y_train_mapped, 3)

# ----------------------------------------------------
# 2. Morphological Opening (Image Processing)
# ----------------------------------------------------
kernel = np.ones((3, 3), np.uint8)

def morphological_opening(img_batch):
    """
    Perform morphological opening on a batch of images.
    Expects shape (batch_size, 32, 32, 3).
    """
    output = []
    for img in img_batch:
        # Convert from float32 [0,1] -> uint8 [0,255]
        temp = (img * 255).astype(np.uint8)
        opened_channels = []
        for ch in cv2.split(temp):
            opened_ch = cv2.morphologyEx(ch, cv2.MORPH_OPEN, kernel)
            opened_channels.append(opened_ch)
        opened = cv2.merge(opened_channels)
        # Convert back to float32 [0,1]
        opened = opened.astype('float32') / 255.0
        output.append(opened)
    return np.array(output, dtype='float32')

# Apply morphological opening on x_train
x_train = morphological_opening(x_train)

# ----------------------------------------------------
# 3. Build & Train a CNN
# ----------------------------------------------------
def create_model():
    inputs = layers.Input(shape=(32, 32, 3))
    x = layers.Conv2D(64, (3, 3), padding='same', activation='relu')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(64, (3, 3), padding='same', activation='relu')(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Conv2D(128, (3, 3), padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(128, (3, 3), padding='same', activation='relu')(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Dropout(0.4)(x)

    x = layers.Conv2D(256, (3, 3), padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(256, (3, 3), padding='same', activation='relu')(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Dropout(0.5)(x)

    x = layers.Flatten()(x)
    features = layers.Dense(512, activation='relu')(x)
    x = layers.BatchNormalization()(features)
    x = layers.Dropout(0.5)(x)

    outputs = layers.Dense(3, activation='softmax')(x)

    model = models.Model(inputs=inputs, outputs=outputs)
    feature_model = models.Model(inputs=inputs, outputs=features)
    return model, feature_model

model, feature_model = create_model()
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Split for validation
val_indices = np.random.choice(len(x_train), size=int(0.2 * len(x_train)), replace=False)
train_indices = np.setdiff1d(np.arange(len(x_train)), val_indices)

x_val, y_val = x_train[val_indices], y_train_oh[val_indices]
x_train_aug, y_train_aug = x_train[train_indices], y_train_oh[train_indices]

# Data augmentation
datagen = ImageDataGenerator(rotation_range=30,
                             width_shift_range=0.1,
                             height_shift_range=0.1,
                             horizontal_flip=True)
datagen.fit(x_train_aug)

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

model.fit(
    datagen.flow(x_train_aug, y_train_aug, batch_size=64),
    epochs=10,
    validation_data=(x_val, y_val),
    callbacks=[early_stop],
    verbose=1
)

# ----------------------------------------------------
# 4. Feature Extraction & UMAP
# ----------------------------------------------------
train_features = feature_model.predict(x_train_aug)  # shape (N, 512)
scaler = StandardScaler()
train_features_scaled = scaler.fit_transform(train_features)

umap_model = umap.UMAP(n_components=50, n_neighbors=5, min_dist=0.01, random_state=42)
train_umap = umap_model.fit_transform(train_features_scaled)
print("UMAP output shape:", train_umap.shape)

# ----------------------------------------------------
# 5. HDBSCAN Clustering & Outlier Detection
# ----------------------------------------------------
# HDBSCAN automatically labels some points as outliers (label = -1).
import hdbscan

# 1) Create HDBSCAN clusterer with 'prediction_data=True'
clusterer = hdbscan.HDBSCAN(
    min_cluster_size=20,
    min_samples=10,
    metric='euclidean',
    cluster_selection_epsilon=0.0,
    gen_min_span_tree=True,
    prediction_data=True    # <-- crucial for approximate prediction
)

# Fit your training data (the UMAP embedding)
cluster_labels = clusterer.fit_predict(train_umap)

# 'probabilities_' gives cluster membership confidence
# 'outlier_scores_' also available for further analysis
outlier_scores = clusterer.outlier_scores_

print("Unique labels assigned by HDBSCAN:", np.unique(cluster_labels))

# ----------------------------------------------------
# 6. Map Each Cluster to Its Most Common True Class
# ----------------------------------------------------
non_outlier_mask = (cluster_labels != -1)
cluster_to_class = {}

if np.any(non_outlier_mask):
    for c_id in np.unique(cluster_labels[non_outlier_mask]):
        if c_id == -1:
            continue
        idxs = np.where(cluster_labels == c_id)[0]
        # Map to majority label
        true_labels = y_train_mapped[train_indices][idxs]
        most_common_label = Counter(true_labels).most_common(1)[0][0]
        class_names = {0: 'airplane', 1: 'automobile', 2: 'ship'}
        cluster_to_class[c_id] = class_names[most_common_label]
else:
    # fallback if everything is outlier
    cluster_to_class = {}

print("Cluster to Class Mapping (excluding -1):", cluster_to_class)

# ----------------------------------------------------
# 7. Classification + Novelty (No Distance Checks)
#    We rely on HDBSCAN cluster label = -1 as "novel".
# ----------------------------------------------------

def classify_array(img):
    # Morphological opening
    img_proc = morphological_opening(img[np.newaxis, ...])[0]

    # CNN -> features -> scale -> UMAP
    feats = feature_model.predict(img_proc[np.newaxis, ...])
    feats_scaled = scaler.transform(feats)
    feats_umap = umap_model.transform(feats_scaled)

    # d) HDBSCAN approximate prediction
    label_array, strengths = approximate_predict(clusterer, feats_umap)
    label = label_array[0]

    # e) Check if label = -1 or not in cluster_to_class => novel
    if label == -1 or label not in cluster_to_class:
        return "Novel class detected", label
    else:
        return f"Classified as: {cluster_to_class[label]}", label
# ----------------------------------------------------
# 8. Test on CIFAR-10 Classes
# ----------------------------------------------------
class_label_to_name = {
    0: 'airplane',
    1: 'automobile',
    2: 'bird',
    3: 'cat',
    4: 'deer',
    5: 'dog',
    6: 'frog',
    7: 'horse',
    8: 'ship',
    9: 'truck'
}

def test_on_random_images(num_samples_per_class=5):
    for c_id in range(10):
        indices = np.where(y_test_full.squeeze() == c_id)[0]
        if len(indices) < num_samples_per_class:
            random_indices = indices
        else:
            random_indices = np.random.choice(indices, num_samples_per_class, replace=False)

        print(f"\n=== True Class {c_id} ({class_label_to_name[c_id]}) ===")
        fig, axes = plt.subplots(1, len(random_indices), figsize=(15,3))
        if len(random_indices) == 1:
            axes = [axes]

        for ax, idx in zip(axes, random_indices):
            img = x_test[idx]
            result, label = classify_array(img)
            ax.imshow(img)
            ax.axis('off')
            color = 'green' if "Classified as" in result else 'red'
            ax.set_title(result, color=color, fontsize=9)
        plt.tight_layout()
        plt.show()

test_on_random_images(num_samples_per_class=5)
```
