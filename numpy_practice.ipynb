##Exercise: Write a function
As an exercise let's write a funciton that displays matrices better, less commas and brackets:

````
def MatPrint(M, message = 'Matrix:'):
  '''
  this fuction accepts a variable M, expected to be in the form of np.array
  and prints it more like a matrix
  Potential future work:
    digit size is currently fixed to 8, therefore it will not show arrays with long numbers properly
  '''
  try:
    if M.ndim != 2: # this function is meant to print only 2D arrays, if not 2D just print it
      print(message)
      print(M)
    else:
      print(message)
      spc = ' '
      sep = '|'
      for r in M: # go over each row, i.e. 1 and 2 :)
        res = sep + spc
        for d in r: # go over each element in the current row
          res= res+ '{:8.2f}'.format(d) + spc
        res = res + sep
        print(res)
  except:
    print(f'Is {M} really a meaningful numpy array?')

a = np.ones((2,2),dtype=int)
b = np.ones((2,3),dtype=int)
c = np.full((2,2,3), 15,dtype=float,)
d = np.zeros((2, 2, 3),dtype=int)
matrix = MatPrint(a)
matrix2 = MatPrint(b)
matrix3 = MatPrint(c)
matrix4 = MatPrint(d)

````

##Operations are elementwise for numpy arrays
Check out the following for:  

Addition, subtraction, multiplication and division are elementwise among numpy arrays

````
a = np.array([1,2,3])
b = np.array([3,2,1])
print(f'a\t= {a}')
print(f'b\t= {b}')
print(f'a+b\t= {a+b}')
print(f'a.b\t= {a*b}')
print(f'a/b\t= {a/b}')
````

No that same deal holds between a numpy array and a scalar:

````
print(f'a\t= {a}')
print(f'2a\t= {a*2}')
print(f'a/2\t= {a/2}')
print(f'a+2\t= {a+2}')
print(f'a-2\t= {a-2}')
````

The same deal holds for functions as well.  
You can pass the whole array through np functions without a loop or anything special for that matter.

For the fun of it check if you can use Eigendecomposition to get the original matrix back  
**Recall:**
For the sake of argument let's focus on a 2x2 case. Given a non-singular 2x2 matrix $M$, the diagonal matrix $\Lambda = \begin{bmatrix} \lambda_1 && 0 \\ 0 && \lambda_2 \end{bmatrix}$ is formed by using the eigenvalues ($\lambda_1, \lambda_2$) of $M$.  
If eigenvectors $v_1, v_2$ corresponding to  $\lambda_1, \lambda_2$ are the columns of matrix $Q$, then we can diagonalize  / decompose $M$ as follows:

> $M = Q \Lambda Q^{-1}$

For more details check out [eigendecomposition](https://en.wikipedia.org/wiki/Eigendecomposition_of_a_matrix) or recall ME210.  

Now let's verify this:


````
A = np.array([[3, 1],[5, 2]])
MatPrint(A, 'Original Matrix A:')
eVals, eVect = np.linalg.eig(A)
eValsD = np.diag(eVals)
MatPrint(eValsD, 'Eigenvalues:')
MatPrint(eVect, 'Corresponding Eigenvectors')
RA = np.matmul(eVect, np.matmul(eValsD, np.linalg.inv(eVect)))
MatPrint(RA, 'Matrix A constructed using Eigenvalues and Eigenvectors')
````

### How do you decide on where to cut?
To determine a single threshold, histograms are very helpful.  
Let's check the histogram of ```homer``` first by converting it into a grayscale iamge using the built in ```rgb2gray``` function from ```skimage.color``` and then using the ```histogram``` function of ```numpy```

````
# convert homer to gray scale
homerG = skimage.color.rgb2gray(homer)
# can be in in [0.0-1.0] or [0-255] range
# above image is in [0.0-1.0] range
# conver to [0-255] range for convenience
homerG255 = np.uint8(homerG*255)
# Histogram.
values, bins = np.histogram(homerG255,bins=np.arange(256))

plt.subplot(2,1,1)
plt.imshow(homerG, cmap=plt.get_cmap("gray"))
plt.subplot(2,1,2)
plt.plot(bins[:255], values)
plt.xlabel('histogram')
````

````
@widgets.interact(x=(0, 255))
def sliderw(x):
  plt.subplot(1,2,1)
  plt.imshow(R, cmap=plt.get_cmap("gray"))
  plt.xlabel('original B channel')
  plt.subplot(1,2,2)
  Bthereshold = x
  plt.imshow(np.where(B <= Bthereshold, 255, 0),cmap=plt.get_cmap("gray"))
  plt.xlabel(f'pixels that are less than {Bthereshold}\n are white, else black ')
````
